```python
import numpy as np
import statistics as sta
import matplotlib.pyplot as plt
#from astropy.table import Table
%pylab inline 
```

```python
Datos = np.genfromtxt("datosregresion.csv", delimiter=",")

Dis = np.array(Datos[:, 0])
Vel = np.array(Datos[:, 1])
```

```python
#Vectores ordenados
SortDis = np.sort(Dis) 
SortVel = np.sort(Vel)
Q_1D = np.percentile(SortDis, 25)
Q_1V = np.percentile(SortVel, 25)
Q_2D = np.percentile(SortDis, 50)
Q_2V = np.percentile(SortVel, 50)
Q_3D = np.percentile(SortDis, 75)
Q_3V = np.percentile(SortVel, 75)

IQR_D = Q_3D - Q_1D
IQR_V = Q_3V - Q_1V
print(IQR_D)
print(IQR_V)
```

```python
MeanDis = sta.mean(Dis)
MeanVel = sta.mean(Vel)
print("La media de distancia:", MeanDis,
      "\nLa media de velocidad:", MeanVel)
'''
ModeDis = sta.mode(Dis)
ModeVel = sta.mode(Vel)
print("La moda de Distancia:", ModeDis,
      "\nLa moda de Velocidad: No tiene",ModeVel)
'''
MedianDis = sta.median(Dis)
MedianVel = sta.median(Vel)
print("La mediana de Distancia:", MedianDis,
      "\nLa mediana de Velocidad:", MedianVel)

StdevDis = sta.stdev(Dis)
StdevVel = sta.stdev(Vel)
print("La desviacion estandar de Distancia:", StdevDis,
      "\nLa desviacion estandar de Velocidad:", StdevVel)

VariDis = sta.variance(Dis)
VariVel = sta.variance(Vel)
print("La varianza de Distancia:", VariDis,
      "\nLa varianza de Velocidad:", VariVel)
```
